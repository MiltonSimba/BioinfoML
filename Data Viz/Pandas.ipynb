### Pandas (NaN)

- Missing values

```python
A = pd.DataFrame([{'a':1,"b":2},{'b':3,'c':4}])
```

```python
A
```

#### Handling missing values

#### 1. Dropping missing values

```python
A.dropna()
#this drops all rows containing Nan
#quite a compromise
```

#### 2. Filling with another number

```python
A.fillna(0)
#sometimes you just have to fill the columns with 0s
```

- There are other functions on how to work with missing values

#### 3. Checking with IsNull (Boolean)

```python
A.isnull()
# checks which columns have Nan
```

```python
A['b'].isnull()
#no Nan
```

```python
A['c'].isnull()
```

 ### Pandas (Grouping)

 - df = pd.DataFrame({'key':[['A','B','C','A','B','C']],'values':range(6)},columns=[['key','values']])   

 - df.groupby('key').sum()

```python
df = pd.DataFrame({'key':list('ABCABC'), 'val':np.arange(6)},columns=['key','val'])
df
```

```python
df.groupby('key').sum() 
```

- groupby function splits a data frame into constituent keys with the same name then in the above example calculates the sum of the grouped keys, and finally creating a new dataframe containing the sum of the numbers.
