### Multiple Plots

- Many plots on a single plot graph.

```python
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1 + y2
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)



```

### Plot Lines Colors

- you can specify your color of  choice for your plots using:
1. Full color's name
2. Color initials
3. RGB values between 0 and 1


```python
plt.plot(x,y1,color="blue")     #full color's name
plt.plot(x,y2, color = "k")     #color initials  'b = blue, k = black
plt.plot(x,y3, color = (1,0.2,0.5))   #RGB values between 0 and 1  
```

### Line Styles

- You can specify the line styles you prefer.
> - Using fullnames of linestyles or   
> - using symbols as short notation (uncommented in the examples below)

```python
# plt.plot(x,x+1,linestyle = 'solid')
plt.plot(x,x+1,linestyle = '-') 

#short notation above
```

```python
# plt.plot(x,y2,linestyle = 'dashdot')
plt.plot(x,y2,linestyle = '-.')

```

```python
# plt.plot(x,x+1,linestyle = 'dashed')

plt.plot(x,x+1,linestyle = '--')

 
```

```python
# plt.plot(x,x+1,linestyle = 'dotted', color = 'r')

plt.plot(x,x+1,linestyle = ':',  color = 'r')

#you can additionally add line colors

```

- Color + line style can be written in shorthand

```python
plt.plot(x,(((x+1)/5)-1.5),':r')
plt.plot(x,y1,'-.b')
plt.plot(x,y2,'-k')
plt.plot(x,y3,'--g')
# multiple plots
# dotted linestyle and red in color
# interesting! huh
```

 ### Putting limits in your plots

```python
plt.plot(x,(((x+1)/5)-1.5),':r')
plt.plot(x,y1,'-.b')
plt.plot(x,y2,'-k')
plt.plot(x,y3,'--g')
plt.xlim(-2,12)
plt.ylim(-2,2)
```

- specifying limits starting with a bigger number flips the graph.

### Additional Functions

```python
plt.plot(x,y1,'-k')
```

### Inverse Graph

- placing limits starting with the bigger number results in plotting of the inverse gragh w.r.t the specified axis

```python
plt.plot(x,y1,'-k')
plt.ylim(2,-2)
```

### Tight function

- values in the x and y axes should be tightened to fit the given window

```python
plt.plot(x,y1,'-k')
plt.axis('tight')

```

### Equal Function

- x and y should be plotted to match the screen aspect ratio

```python
plt.plot(x,y1,'-k')
plt.axis('equal')

```

```python
plt.plot(x,y1,'-k',label = "sin(x)")
plt.plot(x,y2,':g',label = "cos(x)")
plt.plot(x,y3,'-.r',label = "sin(x)+cos(x")
plt.axis('tight')
plt.title('MiltonSimba Plots')
plt.xlabel ('X')
plt.ylabel ('sin(x),cos(x),sin(x)+cos(x)')
plt.legend()


```
