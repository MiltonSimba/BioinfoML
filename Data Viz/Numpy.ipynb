### NumPy Slicing

- uses the notation [[start:end:step]]

```python
# [1:5]   index 1 till 5 but not 5
# [:5]    index 0 till 5 but not 5
# [2:]    index 2 till end including last element
# [::-1]  from end till start (reverse the array)
# [::2]   from start till end every other element
# [::2,:] from every other row and all columns
```

```python
import numpy as np 
```

```python
A = np.random.permutation (np.arange(10))
A
```

```python
A[1:5]
```

```python
A[1:7:2]
```

```python
A[::-1]    #reverses the array
```

```python
A = np.arange(20).reshape(5,4)
```

```python
A
```

```python
A[2,3]    #row with index 2 and element in column with index 3
```

```python
 A[1:3,2]      #from rows with index 1 and 2 access column elements with index 2
```

```python
  A[1:3,1:3]   #from rows with index 1 to 3 access elements in columns  with index 1 to 3
```

```python
A[1:3,:]   #from rows with index 1 to 3 select all elements in the columns
```

```python
A[::2,:]  #from every other elements in all rows select all elements in all the columns
```

### Copy and View

- When you create an array of indices a copy of the array is made which contains elements

```python
A = np.random.permutation(np.arange(10))
A
```

```python
A[[1,5,2]]
```

```python
A<5
```

```python
idx = A<5
A[idx]
#it generates elements rather than boolean
```

```python
A[A<5]   
#can be accessed without calling a variable
```

```python
A[(A<5) & (A>2)]
```

```python
A
```

```python
B = A[1:5]
B      #B is a subset of A
```

```python
B[2]= -100
B
#changed element with index 2
```

```python
A
# a change in array B causes a change in array A
```

```python
B[2]= 3
B
```

```python
A
```

```python
B = A[1:5].copy()
B
#now B is a copy of A rather than a subset of A, any changes in B does not change A
```

```python
B = A[A<5]
B
#by default this is a copy
```

```python
B[0] = 999
B
```

```python
A
#no change happened
```

### Behavior of Numpy

- If you use index, then you dont get a copy
