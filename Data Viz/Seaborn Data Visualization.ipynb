- Seaborn is a high level graphics library built on top of Matplotlib. 


- It is worth noting that seaborn does not bring new functionalities in plotting graphs but rather gives a high level way of plotting graphs compared to Matplotlib. (its simpler)   


- Seaborn and Matplotlib can be used in combination to provide beuatiful visualisation plots.

## Density and pairplots from seaborn


```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import axes3d
```

```python
plt.style.use('classic')
```

```python
x = np.linspace(0,100,1000)
y = np.cumsum(np.random.randn(1000,5),0)

#cumsum = cumulative sum
```

### Histograms

```python
X = 50*np.random.randn(5000)      #first number is the standard deviation
Y = 200*np.random.randn(5000)
Z = 100*np.random.randn(5000)+500
plt.hist(X,density=True,alpha=0.5)
plt.hist(Y,density=True,alpha=0.5)
plt.hist(Z,density=True,alpha=0.5)
```

#### Plotting from pandas

```python
data = pd.DataFrame({'A':X,'B':Y,'C':Z})
```

```python
for col in data.columns:
    plt.hist(data[col],density=True,alpha=0.5)
    sns.kdeplot(data[col],shade = True)

plt.legend ('ABC',ncol = 1)
```

#### Distance Plots

-draws cooler graphs with few lines of code.

```python
for col in data.columns:
    sns.distplot(data[col])
```

#### Joint Plots

- When you want to compare two variables in a DataFrame under the same plot.

```python
with sns.axes_style('white'):
    sns.jointplot('A','B',data,kind='kde')
```

## 3D Plots

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')
r = np.linspace(-2,2,100)**2+1
theta = np.linspace(-4*np.pi,4*np.pi,100)
X = r*np.sin(theta)
Y = r*np.cos(theta)
Z = np.linspace(-2,2,100)
ax.plot(X,Y,Z,label = 'MiltonSimba Made IT')

ax.legend()
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

#### Multiple Plots

```python
sns.set()
fig = plt.figure()
ax = fig.gca(projection='3d')
n = 100
X = (32-23)*np.random.rand(n)+23
Y = (100)*np.random.rand(n)
Z = (-25+50)*np.random.rand(n)-50
ax.scatter(X,Y,Z,c = 'r', marker='o', s=100)
X = (32-23)*np.random.rand(n)+23
Y = (100)*np.random.rand(n)
Z = (-25+50)*np.random.rand(n)-50
ax.scatter(X,Y,Z,c = 'k', marker='^', s=100)
```

### 3d Surface Plot

```python
plt.style.use('bmh')
from mpl_toolkits.mplot3d import axes3d
fig = plt.figure()
ax = fig.gca(projection='3d')
X,Y,Z = axes3d.get_test_data(0.05)
ax.plot_surface(X,Y,Z,cmap='coolwarm', alpha = 1)
```
